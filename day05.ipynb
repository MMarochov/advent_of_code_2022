```python
import re
```

```python
input = open('input_data/day05.txt', 'r').read()
stack_data, moves = input.split("\n\n")
stack_data.splitlines()
moves.splitlines()

print(stack_data)
```

```python
# Get number of stacks
stacks_reversed = stack_data[::-1]
num_of_stacks = int((stacks_reversed[1].split())[0])
num_of_stacks
```

```python
stacks_reversed = stacks_reversed.splitlines()
stacks_reversed = [i[::-1] for i in stacks_reversed]
stacks_reversed
```

```python
stacks = [[] for num in range(num_of_stacks)]
n = 4
for row in stacks_reversed[1:]:
    for i in range(0, len(row), n):
        box = row[i:i+n]
        box = box.replace(' ', '') 
        stack = stacks[int(i/4)]
        if box != '':
            stack.append(box)
stacks
```

```python
moves = moves.splitlines()
move_nums = [[int(s) for s in re.findall(r'\b\d+\b', move)] for move in moves]
move_nums
```
